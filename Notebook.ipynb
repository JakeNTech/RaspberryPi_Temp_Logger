```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20,10)
print("Imported Modules!")
```

We now want to open our data CSV file

```python
df = pd.read_csv("./Sample_Data/test_2.csv")
df.head()
```

Yay! The data is loaded, to make the time stamps easier to work with, i'm going to just convert them to numbers, mostly because I don't know how to make them work nicely with graphs!

```python
df_time = list(range(0,len(df["Date/Time"])))
```

## Visualisations
    1- Temprature Over Time with Average-> Line Graph
    2- Humidity Over Time -> Line Graph
    3- Temperature Over time with linear regresssion

```python
plt.plot(df_time,df["Temperature(S1)"],label="Sensor 1")
plt.plot(df_time,df["Temperature(S2)"],label="Sensor 2")
plt.plot(df_time,df["Average_Temp"],label="Average Temp")
plt.title("Temprature over time")
plt.xlabel("Time")
plt.ylabel("Temprature(C)")
plt.legend(title="Legend")
plt.grid()
```

```python
plt.plot(df_time,df["Humidity(S1)(%)"],label="Sensor 1")
plt.plot(df_time,df["Humidity(S2)(%)"],label="Sensor 2")
plt.plot(df_time,df["Average_Humidity"],label="Average_Humidity")
plt.title("Humidity")
plt.xlabel("Time")
plt.ylabel("Humidity(%)")
plt.legend(title="Legend")
plt.grid()
```

```python
def error_calc(f,x,y):
    return np.sum((f(x)-y)**2)
```

```python
# Bytes Over Time
plt.plot(df_time,df["Temperature(S1)"],label="Sensor 1")
for i in range(1,5):
    poly = np.poly1d(np.polyfit(df_time,df["Temperature(S1)"],i))
    new_x = np.linspace(df_time[0],df_time[-1])
    new_y = poly(new_x)
    plt.plot(new_x, new_y,label="Polynomial %i" %i)
    print(f"The error is: {error_calc(poly,df_time,df['Temperature(S1)'])} for {i}")
#plt.ylim(bottom=0)
plt.xlabel('Time')
plt.ylabel('Temperature(C)')
plt.title("Temperature Over time with regression")
plt.legend(title="Legend")
```
